```python
import matplotlib.pyplot as plt

# 지역별 범죄 발생 횟수의 합 계산
crime_counts_by_region = df.iloc[:, 2:].sum()

# 상위 10개 지역 추출
top_10_regions = crime_counts_by_region.nlargest(10)

# 그래프 그리기
plt.figure(figsize=(10, 6))
top_10_regions.plot(kind='bar', color='skyblue')
plt.title('Top 10 Regions by Crime Counts')
plt.xlabel('Region')
plt.ylabel('Crime Counts')
plt.xticks(rotation=45)
plt.show()
```

```python
# '서울' 열을 기준으로 데이터를 정렬합니다.
sorted_data_seoul = df.sort_values(by='서울', ascending=False)

# 상위 8개의 범죄 중분류를 선택합니다.
top_crime_categories_seoul = sorted_data_seoul['범죄중분류'].head(10)
top_crime_counts_seoul = sorted_data_seoul['서울'].head(10)

# 그래프를 그립니다.
plt.figure(figsize=(10, 7))
plt.bar(top_crime_categories_seoul, top_crime_counts_seoul, color='pink')
plt.xlabel('범죄 중분류')
plt.ylabel('서울 범죄 수')
plt.xticks(rotation=45)
plt.title('서울의 범죄수가 높은 상위 10개 범죄 중분류')
plt.show()
```
